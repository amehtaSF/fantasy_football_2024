````python

url = "https://www.cbssports.com/fantasy/football/stats/{pos}/2024/restofseason/projections/nonppr/"
positions = ["QB", "RB", "WR", "TE", "K"]
def clean_name_cbs(player_str):
    
    # Remove position and anything before it
    player_str =  re.sub(r"^[A-Z]\.(?: St\.)? [a-zA-Z\-']+(?: [a-zA-Z\.]+)?[ ]+(QB|RB|WR|DST|K|TE|FB)", '', player_str).strip()
    # Remove team
    player_str = re.sub(r'\b^[A-Z]{2,3}\b', '', player_str).strip()
    # Remove team from the end
    player_str = re.sub(r' \b[A-Z]{2,3}\b$', '', player_str).strip()
    # Remove position from the end
    player_str = re.sub(r'\b(QB|RB|WR|DST|K|TE|FB)$\b', '', player_str).strip()
    
    return player_str.strip()

def get_team(text):
    re_match = r'(QB|RB|WR|DEF|K|TE|FB)[ ]+[A-Z]{2,3}'
    # extract the match
    match = re.search(re_match, text)
    if not match:
        raise ValueError(f"Could not find team in {text}")
    text = match.group(0)
    text = re.sub(r'(QB|RB|WR|DEF|K|TE|FB)', '', text).strip()
    return text

rename_dict = {
    'Player': 'player_label',
    'Team': 'team',
    'gp  Games Played': 'gp',
    
    'att  Pass Attempts': 'pass_att',
    'cmp  Pass Completions': 'pass_cmp',
    'yds  Passing Yards': 'pass_yd',
    'yds/g  Passing Yards Per Game': 'pass_yd_per_game',
    'td  Touchdowns Passes': 'pass_td',
    'int  Interceptions Thrown': 'pass_int',
    'rate  Passer Rating': 'pass_rate',
    
    'att  Rushing Attempts': 'rush_att',
    'yds  Rushing Yards': 'rush_yd',
    'avg  Average Yards Per Rush': 'rush_yd_per_rsh',
    'td  Rushing Touchdowns': 'rush_td',
    
    'tgt  Targets': 'rec_tgt',
    'rec  Receptions': 'rec_rcpt',
    'avg  Average Yards Per Reception': 'rec_yd_per_rcpt',
    'td  Receiving Touchdowns': 'rec_td',
    'yds  Receiving Yards': 'rec_yd',
    'yds/g  Yards Per Game': 'rec_yd_per_game',
    
    'fgm  Field Goals Made': 'fg_made',
    'fga  Field Goal Attempts': 'fg_att',
    'lng  Longest Field Goal': 'fg_longest',
    '1-19  Field Goals 1-19 Yards': 'fg_1_19_made', 
    '1-19a  Field Goals 1-19 Yard Attempts': 'fg_1_19_att',
    '20-29  Field Goals 20-29 Yards': 'fg_20_29_made',
    '20-29a  Field Goals 20-29 Yard Attempts': 'fg_20_29_att',
    '30-39  Field Goals 30-39 Yards': 'fg_30_39_made',
    '30-39a  Field Goals 30-39 Yard Attempts': 'fg_30_39_att',
    '40-49  Field Goals 40-49 Yards': 'fg_40_49_made',
    '40-49a  Field Goals 40-49 Yard Attempts': 'fg_40_49_att',
    '50+  Field Goals 50+ Yards': 'fg_50_made',
    '50+a  Field Goals 50+ Yards Attempts': 'fg_50_att',
    'xpm  Extra Points Made': 'xp_made',
    'xpa  Extra Points Attempted': 'xp_att',
    
    'Team': 'team', 
    'int  Interceptions': 'def_int', 
    'sfty  Safeties': 'def_sfty', 
    'sck  Sacks': 'def_sack',
    'tk  Tackles': 'def_tkl', 
    'frec  Defensive Fumbles Recovered': 'def_fum_rec',
    'fum  Forced Fumbles': 'def_fum_forced', 
    'dtd  Defensive Touchdowns': 'def_td',
    'pts  Points Allowed': 'def_pts_allowed', 
    'ppg  Points Allowed Per Game': 'def_pts_allowed_per_game',
    'pass  Net Passing Yards Allowed': 'pass_yd_allowed',
    'rush  Rushing Yards Allowed': 'rush_yd_allowed',
    'total  Total Yards Allowed': 'total_yd_allowed',
    'avg  Yards Against Per Game': 'total_yd_allowed_per_game',
    
    'fl  Fumbles Lost': 'fumbles_lost',
    'fpts  Fantasy Points': 'fpts',
    'fppg  Fantasy Points Per Game': 'fp_per_game',
    
}

keep_cols = list(rename_dict.keys())

df_pos = {}
for position in positions:
    
    df_scrape = pd.read_html(url.format(pos=position))[0]
    df_scrape.columns = df_scrape.columns.droplevel()
    # print(f"{position} df_scrape columns: {df_scrape.columns}")
    df = df_scrape.loc[:, [col for col in df_scrape.columns if col in keep_cols]]
    df = df.rename(columns=rename_dict)
    df['pos'] = position
    df['source'] = 'CBS'
    df['source_url'] = url.format(pos=position)
    df['fpts_halfppr_adjust'] = False
    if position in ['RB', 'WR', 'TE']:
        df['fpts'] = df['fpts'] + df['rec_rcpt'] * .5
        df['fpts_halfppr_adjust'] = True
    df['player'] = df['player_label'].apply(clean_name_cbs)
    df['team'] = df['player_label'].apply(get_team)
    df_pos[position] = df
    df.drop(columns=['player_label'], inplace=True)
    df.to_csv(f"data/proj_scrape/cbs/{position}_projections.csv", index=False)
    
# DEF
position = "DST"
df_scrape = pd.read_html(url.format(pos=position))[0]
df_scrape.columns = df_scrape.columns.droplevel()
# print(f"{position} df_scrape columns: {df_scrape.columns}")
df_def = df_scrape.loc[:, [col for col in df_scrape.columns if col in keep_cols]]
# print(f"{position} df_scrape columns: {df_scrape.columns}")
df_def["player"] = df_def["Team"]
df_def = df_def.rename(columns=rename_dict)
df_def['pos'] = "DEF"
df_def['source'] = 'CBS'
df_def['source_url'] = url.format(pos=position)
df_pos["DEF"] = df_def
df_def.to_csv(f"data/proj_scrape/cbs/DEF_projections.csv", index=False)



# df = pd.concat(df_pos.values(), ignore_index=True)
# display(df)

````

````python


# Define the URL template
url = "https://www.fantasysharks.com/apps/bert/forecasts/projections.php?csv=1&Sort=&Position={pos_num}&scoring=18"


rename_dict = {
    "Player Name": "player",
    "Team": "team",
    "Position": "pos",
    
    "Tgt": "rec_tgt",
    "Rec": "rec_rcpt",
    "Rec Yds": "rec_yd",
    "Rec TDs": "rec_td",
    
    "Pass Yds": "pass_yd",
    "Pass TDs": "pass_td",
    
    "Rush": "rush_att",
    "Rush Yds": "rush_yd",
    "Rush TDs": "rush_td",
    
    "XP": "xp_made",
    "XPA": "xp_att",
    "FG": "fg_made",
    "10-19": "fg_1_19_made",
    "20-29": "fg_20_29_made",
    "30-39": "fg_30_39_made",
    "40-49": "fg_40_49_made",
    "50+": "fg_50_made",
    "Miss": "fg_miss",
    
    "YdsAllowed": "yd_allowed",
    "PtsAgn": "pts_allowed",
    "Scks": "def_sack",
    "DefTD": "def_td",
    "Safts": "def_sfty",
    
    
    "Fum Lost": "fumbles_lost",
    "Pts": "fpts"
    
}
pos_rename_dict = {
    "QB": {
        "Att": "pass_att",
        "Comp": "pass_cmp",
        "Int": "pass_int",
        "Sacked": "pass_sacked"
        },
    "RB": {},
    "WR": {},
    "TE": {},
    "K": {
        "Att": "fg_att"
        },
    "DEF": {
        "Int": "def_int",
        "Fum": "def_fum",
    }
}


# Define the positions dictionary
positions = {
    "QB": 1,
    "RB": 2,
    "WR": 4,
    "TE": 5,
    "K": 7,
    "DEF": 6
}

# Define headers with a user agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Loop through each position
for position, pos_num in positions.items():
    # Fetch the CSV data with headers
    response = requests.get(url.format(pos_num=pos_num), headers=headers)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to fetch data for position {position}: {response.status_code}")
        continue
    # Read the content of the response as a CSV
    csv_data = StringIO(response.text)
    df = pd.read_csv(csv_data)
    
    df = df.loc[:, [col for col in df.columns if col in rename_dict.keys() or col in pos_rename_dict[position].keys()]]
    df = df.rename(columns=rename_dict)
    df = df.rename(columns=pos_rename_dict[position])
    # Add additional columns
    df["pos"] = position
    df["source"] = "FantasySharks"
    df["source_url"] = url.format(pos_num=pos_num)
    df["player"] = df["player"].apply(lambda x: x.split(",")[1].strip() + " " + x.split(",")[0].strip())
    
    # Adjust names
    if position == "WR":
        df["player"] = df["player"].apply(lambda x: re.sub(r'D\.J\. Chark', 'DJ Chark', x))
    
    # Save the DataFrame to a CSV file
    df.to_csv(f"data/proj_scrape/fantasysharks/{position}_projections.csv", index=False)

````

````python
url = "https://www.numberfire.com/nfl/fantasy/remaining-projections/{pos}"

positions = ["qb", "rb", "wr", "te", "k", "d"]

def clean_name(name):
    # First remove the parenthetical
    # name = re.sub(r'\s*\([A-Z]{2}, [A-Z]{2,3}\)$', '', name)
    name = re.sub(r'\s*\([A-Z]{2,3}, [A-Z]{2,3}\)$', '', name).strip()
    # Then remove up to the uppercase letter, period, and space
    # name = re.sub(r'\s*[A-Z]\.\s.*$', '', name)
    # name = re.sub(r'\s*[A-Z](?:\.[A-Z])?\.\s.*$', '', name)
    name = re.sub(r"[A-Z]\.\s['\-A-Za-z]+$", '', name).strip()
    name = re.sub(r"Amon\-Ra St\. Brown  A\. St\. Brown", 'Amon-Ra St. Brown', name).strip()
    name = re.sub(r"Michael Pittman Jr\.  M\. Pittman Jr\.", 'Michael Pittman Jr.', name).strip()
    name = re.sub(r"D\.J\. Chark", 'DJ Chark', name).strip()
    name = re.sub(r"Terrace Marshall Jr\.", 'Terrace Marshall', name).strip()
    name = re.sub(r"Donald Parham Jr\.  D\. Parham Jr\.", 'Donald Parham Jr.', name).strip()
    name = re.sub(r"Irv Smith Jr\.  I\. Smith Jr\.", 'Irv Smith Jr.', name).strip()
    return name

def get_team(text):
    re_match = r'\((QB|RB|WR|TE|K|D), [A-Z]{2,3}\)$'
    match = re.search(re_match, text)
    if not match:
        raise ValueError(f"Could not find team in {text}")
    text = match.group(0)
    text = re.sub(r'\(', '', text).strip()
    text = re.sub(r'\)', '', text).strip()
    text = text.split(",")[1].strip()
    return text

rename_dict = {
    'numberFire_FP': 'fpts', 
    'numberFire_CI': 'fpts_ci', 
    'Ranks_Ovr.': 'rank_overall', 
    'Ranks_Pos.': 'rank_pos',
    
    'Passing_C/A': 'pass_cmp_att', 
    'Passing_Yds': 'pass_yd', 
    'Passing_TDs': 'pass_td',
    'Passing_Ints': 'pass_int',
    
    'Rushing_Att': 'rush_att',
    'Rushing_Yds': 'rush_yd',
    'Rushing_TDs': 'rush_td',
    
    'Receiving_Rec': 'rec_rcpt',
    'Receiving_Tgt': 'rec_tgt',
    'Receiving_Yds': 'rec_yd',
    'Receiving_TDs': 'rec_td',
    
    'Kicking_XPM': 'xp_made',
    'Kicking_FGA': 'fg_att',
    'Kicking_FGM': 'fg_made',
    'FG Made By Distance_0-19': 'fg_made_0_19',
    'FG Made By Distance_20-29': 'fg_made_20_29', 
    'FG Made By Distance_30-39': 'fg_made_30_39',
    'FG Made By Distance_40-49': 'fg_made_40_49', 
    'FG Made By Distance_50+': 'fg_made_50_plus',
    
    'Defense_Points Allowed': 'def_pts_allowed', 
    'Defense_Yards Allowed': 'def_yd_allowed', 
    'Defense_Sacks': 'def_sack',
    'Defense_INTs': 'def_int', 
    'Defense_Fumbles': 'def_fum', 
    'Defense_TDs': 'def_td'
}
pos_rename_dict = {
    "qb": "QB",
    "rb": "RB",
    "wr": "WR",
    "te": "TE",
    "k": "K",
    "d": "DEF"
}

for pos in positions:
    print(f"Scraping: {pos}")
    tables = pd.read_html(url.format(pos=pos))
    df_scrape = tables[1]
    df_scrape.columns = ['_'.join(col).strip() for col in df_scrape.columns.values] # flatten the multiindex
    names = [tables[0].columns[0]] + tables[0].iloc[:, 0].tolist() # player names
    
    df = df_scrape.loc[:, [col for col in df_scrape.columns if col in rename_dict.keys()]]
    df = df.rename(columns=rename_dict)

    df["player"] = names
    df["team"] = df["player"].apply(get_team)
    df["player"] = df["player"].apply(clean_name)
    
    
    df["pos"] = pos
    df["source"] = "NumberFire"
    df["source_url"] = url.format(pos=pos)
    df["fpts_halfppr_adjust"] = False
    if pos in ["rb", "wr", "te"]:
        df["fpts"] = df["fpts"] + df["rec_rcpt"] * .5
        df["fpts_halfppr_adjust"] = True
    if pos == "qb":
        df["pass_cmp"] = df["pass_cmp_att"].apply(lambda x: float(str(x).split("/")[0]))
        df["pass_att"] = df["pass_cmp_att"].apply(lambda x: float(str(x).split("/")[1]))
        df.drop(columns=["pass_cmp_att"], inplace=True)
    pos_label = pos_rename_dict[pos]
    df.to_csv(f"data/proj_scrape/numberfire/{pos_label}_projections.csv", index=False)
````

```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from pprint import pprint
```

```python
# https://sleeper-api-client.readthedocs.io/en/latest/endpoints/stats.html
# https://api.sleeper.com/stats/nfl/<season>[/<week>]?season_type=<season_type>[&position=<position>]

url = "https://api.sleeper.com/stats/nfl/2022?season_type=reg&position=QB"

resp = requests.get(url)

if resp.status_code == 200:
    pprint(resp.json())
else:
    print("error", resp.status_code, resp.text)
```

```python
# https://api.sleeper.com/stats/nfl/player/<player_id>?season=<season>&season_type=<season_type>&grouping=week

url = "https://api.sleeper.com/stats/nfl/player/4984?season=2022&season_type=regular&grouping=week"

resp = requests.get(url)

if resp.status_code == 200:
    pprint(resp.json())
else:
    print("error", resp.status_code, resp.text)
```

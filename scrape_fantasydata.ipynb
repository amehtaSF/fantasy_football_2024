```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

# Get all weeks


```python

driver = webdriver.Chrome()
url = "https://fantasydata.com/nfl/fantasy-football-leaders?scope=game&scoring=fpts_half_ppr&order_by=team&sort_dir=asc"

try:
    driver.get(url)
    wait = WebDriverWait(driver, 3)
    dropdown = wait.until(EC.presence_of_element_located((By.NAME, 'spw')))
    options = dropdown.find_elements(By.TAG_NAME, 'option')
    weeks = []
    for option in options:
        week = option.text
        print(week)
        weeks.append(week)

finally:
    driver.quit()

# Week takes the form "spw=2023_REG_10"
```

# Get all teams

```python

driver = webdriver.Chrome()
url = "https://fantasydata.com/nfl/fantasy-football-leaders?scope=game&scoring=fpts_half_ppr&order_by=team&sort_dir=asc"

try:
    driver.get(url)
    wait = WebDriverWait(driver, 3)
    dropdown = wait.until(EC.presence_of_element_located((By.NAME, 'team')))
    options = dropdown.find_elements(By.TAG_NAME, 'option')
    teams = []
    for option in options:
        team = option.text
        print(team)
        teams.append(team)

finally:
    driver.quit()
```
